```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import handytools.navigator as nav

def plot_img(*files):
    image_path = os.path.join("/home2/belleliu/artiphisiology/graphs/", *files)
    image = mpimg.imread(image_path)
    fig = plt.figure(dpi=150)
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

Below is the FOI of different networks across layers. The left panel shows the mean, and the right panel shows a box plot for its distribution.

## GPB images for top and bottom FOI units

The unit number of the top FOI (corresponds to small TI) and bottom FOI (large TI) units are shown below. To see the corresponding guided backpropagation images, enter the network type (AN, VGG16b, ResNet18) and unit number in the function ``show_gbp``. To see high resolution images, go to

``/home2/belleliu/artiphisiology/graphs/gbp_{network type}/``

to find the images. For Alexnet, look for layer Conv3 (``hkey=6``); for VGG16, look for layer Conv9 (``hkey=19``); for ResNet18, look for basic block 3.1 (``hkey=8``).

CAUTION: Some of the results are still running and may not be present.

```python
units = nav.npload(nav.resultpath, f"responses_AN", f"hkey=6_unitsTiMin.npy")
print("High FOI, low TI units in AlexNet, Conv3: ", ", ".join([str(u) for u in units]))

units = nav.npload(nav.resultpath, f"responses_AN", f"hkey=6_unitsTiMax.npy")
print("Low FOI, High TI units in AlexNet, Conv3: ", ", ".join([str(u) for u in units]))

units = nav.npload(nav.resultpath, f"responses_VGG16b", f"hkey=19_unitsTiMin.npy")
print("High FOI, low TI units in VGG16, Conv9: ", ", ".join([str(u) for u in units]))

units = nav.npload(nav.resultpath, f"responses_VGG16b", f"hkey=19_unitsTiMax.npy")
print("Low FOI, High TI units in VGG16, Conv9: ", ", ".join([str(u) for u in units]))

units = nav.npload(nav.resultpath, f"responses_ResNet18", f"hkey=8_unitsTiMin.npy")
print("High FOI, low TI units in ResNet18, b-block3.1: ", ", ".join([str(u) for u in units]))

units = nav.npload(nav.resultpath, f"responses_ResNet18", f"hkey=8_unitsTiMax.npy")
print("Low FOI, High TI units in ResNet18, b-block3.1: ", ", ".join([str(u) for u in units]))
```

```python
def show_gbp(mtype, unit, tpe="top"):
    hdic = {"AN": 6, "VGG16b": 19, "ResNet18": 8}
    plot_img(f"gbp_{mtype}", f"hkey={hdic[mtype]}_unit={unit}_{tpe}.png")
```

```python
show_gbp("AN", 57)
```
